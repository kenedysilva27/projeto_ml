```python
numeric_cols = [col for col in df_train.columns if col.startswith('var_')]

df_train['median_features'] = df_train[numeric_cols].median(axis=1)
df_train['std_features'] = df_train[numeric_cols].std(axis=1)
df_train['min_features'] = df_train[numeric_cols].min(axis=1)
df_train['max_features'] = df_train[numeric_cols].max(axis=1)

original_cols = [col for col in df_train.columns if col.startswith('var_')]
total_features = len(original_cols)

df_train['count_positive'] = (df_train[original_cols] > 0).sum(axis=1)
df_train['count_negative'] = (df_train[original_cols] < 0).sum(axis=1)

df_train['prop_positive'] = df_train['count_positive'] / total_features
df_train['prop_negative'] = df_train['count_negative'] / total_features
```

```python
df_train
```

```python
feature_cols = [col for col in df_train.columns if col.startswith('var_')]
X = df_train[feature_cols].copy()
y = df_train['target'].copy()
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import lightgbm as lgb
import time

models = {
    'Logistic Regression': LogisticRegression(
        random_state=42, 
        max_iter=1000,
        C=1.0
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        n_jobs=-1,
        max_depth=10,
        min_samples_split=100
    ),
    'LightGBM': lgb.LGBMClassifier(
        n_estimators=200,
        random_state=42,
        verbose=-1,
        learning_rate=0.1,
        num_leaves=31
    )
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = {}
for name, model in models.items():
    print(f"Testando: {name}")
    start_time = time.time()
    
    scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc', n_jobs=-1)
    
    mean_score = scores.mean()
    std_score = scores.std()
    train_time = time.time() - start_time
    
    results[name] = {
        'mean_auc': mean_score,
        'std_auc': std_score,
        'time': train_time
    }
    
    print(f" AUC: {mean_score:.4f} (±{std_score:.4f}) - {train_time:.1f}s")

best_model = max(results.keys(), key=lambda k: results[k]['mean_auc'])
print(f"\nMelhor modelo: {best_model}")
print(f"AUC: {results[best_model]['mean_auc']:.4f}")
```
